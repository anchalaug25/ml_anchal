```python
df.isnull().mean()*100
```

```python
x=df.drop("PerformanceScore",axis=1)
y=df["PerformanceScore"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
cvs=cross_val_score(lr,x_train,y_train,scoring="r2",cv=5)
```

```python
cvs
```

```python
np.mean(cvs)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
#now we will perform metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,accuracy_scoreS
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
r2
```
