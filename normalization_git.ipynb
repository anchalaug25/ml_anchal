```python
df=pd.read_csv("wine_data.csv",header=None,usecols=[1,2,3])
```

```python
df.head()
```

```python
df.columns=["class_label","alcohol","malic_acid"]
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x=df.drop("class_label",axis=1)
y=df["class_label"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled #these values are in numpy array so we will convert them to dataframe
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
```

```python
x_train_scaled #now they are in dataframe form
```

```python
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```
