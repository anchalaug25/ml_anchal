```python
df=pd.read_csv("Social_Network_Ads.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x=df[["Age","EstimatedSalary"]]
y=df["Purchased"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train_new=sc.transform(x_train)
x_test_new=sc.transform(x_test)
```

```python
x_train_new #these values is in numpy array so we will convert them to dataframe
```

```python
x_train_new=pd.DataFrame(x_train_new,columns=x_train.columns)
x_test_new=pd.DataFrame(x_test_new,columns=x_test.columns)
```

```python
x_train_new # now these values are in dataframe
```

```python
np.round(x_train.describe(),1) #here values are varying a lot like age and estimated there difference is large
```

```python
np.round(x_train_new.describe(),1) #here its gud
```

```python
#applyting standardize in logistic to check accuracy we can apply on any algo
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr_scaled=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_scaled.fit(x_train_new,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred_new=lr_scaled.predict(x_test_new)
```

```python
from sklearn.metrics import accuracy_score
print("actual",accuracy_score(y_test,y_pred))
print("scaled",accuracy_score(y_test,y_pred_new))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
dt_scaled.fit(x_train_new,y_train)
```

```python
y_pred = dt.predict(x_test)
y_pred_scaled = dt_scaled.predict(x_test_new)
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_new))
```
